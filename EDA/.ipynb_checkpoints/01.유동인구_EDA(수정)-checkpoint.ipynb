## Data Load

# 성연령 유동인구

## Data Load

```python
age = eda.flow_concat('AGE','../data/01_SKT/*.csv')
```

## Data Proprocessing

```python
age = eda.flow_preprocessing(age)
print(age.shape)
age.head(3)
```

```python
age = eda.flow_melting(age)
```

```python
age['AGE_CD'] = age['VAR'].apply(lambda x: x[-3:] if x[-4] == '_' else x[-4:])
age['SEX_CD'] = age['VAR'].str[0].map({'M':'M', 'W':'F'})
age = age[['STD_YMD', 'GU_NM', 'HDONG_NM', 'SEX_CD', 'AGE_CD', 'POP_CNT']]
age.head()
```

```python
age_dict = {x: int(x[:2])for x in sorted(age['AGE_CD'].unique())}
age['AGE_CD'] = age['AGE_CD'].map(age_dict)
age.groupby(['AGE_CD'])['POP_CNT'].mean()
```

### comment
0~20세까지의 유동인구가 다른 연령층에 비해 적은 수치임 </p>
20세미만은 20로 포함시키고 65세 이상은 65로 포함시킴

```python
age['AGE_CD'] = age['AGE_CD'].apply(lambda x: 20 if x < 20 else x)
age['AGE_CD'] = age['AGE_CD'].apply(lambda x: 65 if x > 65 else x)
age = age.groupby(['STD_YMD', 'GU_NM', 'HDONG_NM', 'SEX_CD', 'AGE_CD'])[['POP_CNT']] \
         .sum().reset_index()
```

```python
age['DAY_NM'] = age['STD_YMD'].apply(lambda x: x.day_name())
age = age.set_index('STD_YMD')
age.head()
```

## 행정동별 유동인구

```python
palett = ['red', 'orange', 'olive', 'green', 'cyan', 'blue','purple']
year_flow = ['2019', '2020']

fig, ax = plt.subplots(2, 1, figsize=(16, 12), sharey=True, sharex=True)

for year in year_flow:
    age[year].groupby('HDONG_NM')['POP_CNT'].agg('sum').plot.bar(color=['tab:' + str(i) for i in palett], ax=ax[year_flow.index(year)])
    ax[year_flow.index(year)].set_title(f'{year}년 행정동별 유동인구', fontsize = 20)
    ax[year_flow.index(year)].set_xlabel('행정동')
    ax[year_flow.index(year)].set_ylabel('유동인구')

plt.show()
```

## 행정동별 유동인구 변화

```python
# 각 행정동별 연도별 pop_cnt의 합계
age.groupby([pd.Grouper(freq='Y'), 'HDONG_NM'])['POP_CNT'].sum().reset_index().set_index('STD_YMD')
```

```python
sj_age19 = age['2019'].query('GU_NM == "서울 중구"').groupby('HDONG_NM')['POP_CNT'].sum()
sj_age20 = age['2020'].query('GU_NM == "서울 중구"').groupby('HDONG_NM')['POP_CNT'].sum()

sn_age19 = age['2019'].query('GU_NM == "서울 노원구"').groupby('HDONG_NM')['POP_CNT'].sum()
sn_age20 = age['2020'].query('GU_NM == "서울 노원구"').groupby('HDONG_NM')['POP_CNT'].sum()

dj_age19 = age['2019'].query('GU_NM == "대구 중구"').groupby('HDONG_NM')['POP_CNT'].sum()
dj_age20 = age['2020'].query('GU_NM == "대구 중구"').groupby('HDONG_NM')['POP_CNT'].sum()

ds_age19 = age['2019'].query('GU_NM == "대구 수성구"').groupby('HDONG_NM')['POP_CNT'].sum()
ds_age20 = age['2020'].query('GU_NM == "대구 수성구"').groupby('HDONG_NM')['POP_CNT'].sum()
```

```python
#대부분의 유동인구가 감소하였기 때문에 절대값을 사용하여 감소율을 나타냄
flow_variation_sj =abs((sj_age20 - sj_age19)/ sj_age19).reset_index().set_index('HDONG_NM')
flow_variation_sn =abs((sn_age20 - sn_age19)/ sn_age19).reset_index().set_index('HDONG_NM')
flow_variation_dj =abs((dj_age20 - dj_age19)/ dj_age19).reset_index().set_index('HDONG_NM')
flow_variation_ds =abs((ds_age20 - ds_age19)/ ds_age19).reset_index().set_index('HDONG_NM')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(40, 30), sharey = True)

flow_variation_sj.plot.bar(ax=ax[0,0])
ax[0,0].set_title('서울 중구 인구 감소율', fontsize = 30)
plt.setp(ax[0,0].get_xticklabels(), fontsize=20)
ax[0,0].legend('감소율', fontsize = 20)
ax[0,0].set_xlabel('')

flow_variation_sn.plot.bar(ax=ax[0,1])
ax[0,1].set_title('서울 노원구 인구 감소율', fontsize = 30)
plt.setp(ax[0,1].get_xticklabels(), fontsize=20)
ax[0,1].legend('감소율', fontsize = 20)
ax[0,1].set_xlabel('')

flow_variation_dj.plot.bar(ax=ax[1,0])
ax[1,0].set_title('대구 중구 인구 감소율', fontsize = 30)
plt.setp(ax[1,0].get_xticklabels(), fontsize=20)
ax[1,0].legend('감소율', fontsize = 20)
ax[1,0].set_xlabel('')

flow_variation_ds.plot.bar(ax=ax[1,1])
ax[1,1].set_title('대구 수성구 인구 감소율', fontsize = 30)
plt.setp(ax[1,1].get_xticklabels(), fontsize=20)
ax[1,1].legend('감소율', fontsize = 20)
ax[1,1].set_xlabel('')

plt.show()
```

```python
flow_variation_sj = geo.merge(flow_variation_sj, on = 'HDONG_NM')
flow_variation_sn = geo.merge(flow_variation_sn, on = 'HDONG_NM')
flow_variation_dj = geo.merge(flow_variation_dj, on = 'HDONG_NM')
flow_variation_ds = geo.merge(flow_variation_ds, on = 'HDONG_NM')
```

```python
fig, ax = plt.subplots(2, 2, figsize=(30, 20))

flow_variation_sj.plot(column='POP_CNT', cmap = 'Blues', legend=True, ax=ax[0,0])
ax[0,0].set_title('작년 대비 서울 중구 유동인구 감소율', fontsize = 25)
ax[0,0].set_axis_off()

flow_variation_sn.plot(column='POP_CNT', cmap = 'Blues', legend=True, ax=ax[0,1])
ax[0,1].set_title('작년 대비서울 노원구 유동인구 감소율', fontsize = 25)
ax[0,1].set_axis_off()

flow_variation_dj.plot(column='POP_CNT', cmap = 'Blues', legend=True, ax=ax[1,0])
ax[1,0].set_title('작년 대비 대구 중구 유동인구 감소율', fontsize = 25)
ax[1,0].set_axis_off()

flow_variation_ds.plot(column='POP_CNT', cmap = 'Blues', legend=True, ax=ax[1,1])
ax[1,1].set_title('작년 대비 대구 수성구 유동인구 감소율', fontsize = 25)
ax[1,1].set_axis_off()

plt.show()
```

```python
daegu = pd.read_csv('../data/06_Epidemic_Datasets/Daegu_Covid.csv', 
                    encoding='cp949', parse_dates=['날짜'])
```

```python
daegu = daegu.iloc[:, :2]
daegu.rename(columns={'날짜':'STD_YMD', '당일 확진자 수':'PT_CNT'}, inplace=True)
daegu = daegu.set_index('STD_YMD')

for i in range(17):
    daegu.loc[pd.to_datetime('2020-02-01') + pd.DateOffset(days=i)] = [0]
daegu = daegu.sort_values(by='STD_YMD')    
daegu = daegu[:'2020-05']
daegu.head()
```

```python
day_name = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
color = ['blue', 'orange', 'green', 'brown', 'pink', 'purple', 'red']

fig, ax = plt.subplots(2, 1, figsize=(12, 16))

for i in range(len(day_name)):
    age_day = age.query('GU_NM in ["대구 수성구", "대구 중구"] and DAY_NM == @day_name[@i]') \
                 .groupby([pd.Grouper(freq='D'), 'GU_NM', 'HDONG_NM'])[['POP_CNT']].sum().reset_index().set_index('STD_YMD')
    
    if i > 4: age_day = age_day['2020-02-03':]
    age_day['2020'][['POP_CNT']].resample('W').sum() \
                                .plot(figsize=(12, 8), color='tab:' + color[i], ax=ax[0])
    ax[0].legend(day_name)
    ax[0].set_xlabel('날짜')
    ax[0].set_title('대구시 요일별 유동인구 추이', fontsize = 17)


daegu['2020-02-09':].plot(color='red', ax=ax[1])
ax[1].set_title('대구시 확진자 수 추이', fontsize = 17)
ax[1].legend(['확진자'])
ax[1].set_xlabel('날짜')

fig.tight_layout(h_pad=2)

plt.show()
```

## 전년도 동일 대비 유동인구 감소율(POP_RD열)

```python
for year in ['19', '20']:
    globals()[f'daegu_{year}'] = age.query('GU_NM in ["대구 수성구", "대구 중구"]')[f'20{year}'] \
                        .groupby([pd.Grouper(freq='D'), 'HDONG_NM'])['POP_CNT'].sum() \
                        .reset_index().set_index('STD_YMD')
```

```python
rate = pd.merge(daegu_20, daegu_19.set_index(daegu_19.index + pd.DateOffset(years=1)), on=['STD_YMD', 'HDONG_NM'],
                            how='left', suffixes=('_20', '_19')).reset_index()
```

```python
idx = rate[rate['POP_CNT_19'].isnull()].index
hd = rate[rate['POP_CNT_19'].isnull()]['HDONG_NM']

for i, j in zip(idx, hd):
    rate.loc[i, 'POP_CNT_19'] = rate.query('HDONG_NM == @j') \
                                    .query('STD_YMD in ["2020-02-28", "2020-03-01"]') \
                                    .groupby('HDONG_NM')['POP_CNT_19'].mean().values[0]
```

```python
rate['POP_ROD'] = (rate['POP_CNT_20'] - rate['POP_CNT_19']) / rate['POP_CNT_19']

rate = rate.set_index('STD_YMD')
rate = rate[['HDONG_NM', 'POP_ROD']]
rate
```

```python
age_daegu = pd.merge(age_daegu, rate, on=['STD_YMD', 'HDONG_NM'])
age_daegu.head()
```

## 행정동별 평균 유동인구

```python
df = pd.merge(geo[['HDONG_NM', 'AREA', 'LENGTH', 'X_COORD', 'Y_COORD', 'geometry']],
              age, on='HDONG_NM')
```

```python
def temp(gu):
    h_group = df[df['GU_NM'] == gu].groupby('HDONG_NM')['POP_CNT'].mean().reset_index()
    df2 = df[df['GU_NM'] == gu].iloc[:, :7].drop_duplicates(subset='HDONG_NM')
    df2 = pd.merge(df2, h_group, on='HDONG_NM')
    df2 = gpd.GeoDataFrame(df2)
    return df2
```

```python
fig, ax = plt.subplots(2, 2, figsize=(12, 12))

temp('서울 중구').plot(column='POP_CNT', cmap='Blues', ax=ax[0,0])
ax[0,0].set_title('서울 중구 동별 평균 유동인구')
ax[0,0].set_axis_off()

temp('서울 노원구').plot(column='POP_CNT', cmap='Blues', ax=ax[0,1])
ax[0,1].set_title('서울 노원구 평균 유동인구')
ax[0,1].set_axis_off()

temp('대구 중구').plot(column='POP_CNT', cmap='Blues', ax=ax[1,0])
ax[1,0].set_title('대구 중구 동별 평균 유동인구')
ax[1,0].set_axis_off()

temp('대구 수성구').plot(column='POP_CNT', cmap='Blues', ax=ax[1,1])
ax[1,1].set_title('대구 수성구 동별 평균 유동인구')
ax[1,1].set_axis_off()

plt.show()
```
