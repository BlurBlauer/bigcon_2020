```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import eda
import warnings
warnings.filterwarnings(action='ignore')
```

```python
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode, iplot
from plotly.subplots import make_subplots
import geopandas as gpd
import matplotlib.colors as colors
```

```python
md = pd.read_excel('../data/05_CJ_Olive_Networks/cj_dataset.xlsx')
geo = gpd.read_file('../data/01_SKT/4개지역_행정동.shp')
```

```python
#geo에서 필요한 정보만을 추출.
geo = geo[['HDONG_NM', 'SIDO_NM', 'SGNG_NM',
          'AREA', 'LENGTH', 'X_COORD', 'Y_COORD', 'geometry']]
```

```python
def to_datetime(df, col):
    if (df[col].dtypes == np.dtype(np.int32) or df[col].dtypes == np.dtype(np.int64)):
        df[col] = df[col].astype('str').astype('datetime64')        
    elif df[col].dtypes == np.dtype(np.object):
        df[col] = df[col].astype('datetime64')        
    else:
        print('Please check the data type again.')
```

```python
md['DL_YMD'] = pd.to_datetime(md['DL_YMD'], format='%y%m%d')
md['CTGG_NM'] = md['CTPV_NM'].str[:2] + ' '+ md['CTGG_NM']
md.drop(['CTPV_NM','CTPV_CD', 'CTGG_CD', 'HDNG_CD', 'DL_GD_LCLS_CD'], axis=1, inplace=True)
md = md.set_index('DL_YMD')
```

# 카테고리 별 송장 건수 변화

```python
md19 = md['2019']
md20 = md['2020']
```

```python
fig, ax = plt.subplots(2, 1, figsize=(16, 10), sharey= True)

md19.groupby([lambda x: md19.index.month, 'DL_GD_LCLS_NM'])['INVC_CONT'].sum().unstack('DL_GD_LCLS_NM').plot.bar(ax = ax[0])
ax[0].set_title('2019년 카테고리, 월별 누적 송장 건수', fontsize = 20)
md20.groupby([lambda x: md20.index.month, 'DL_GD_LCLS_NM'])['INVC_CONT'].sum().unstack('DL_GD_LCLS_NM').plot.bar(ax = ax[1])
ax[1].set_title('2020년 카테고리, 월별 누적 송장 건수', fontsize = 20)

plt.show()
```

```python
cate_inc = (md20.groupby(['DL_GD_LCLS_NM'])['INVC_CONT'].sum() - md19.groupby(['DL_GD_LCLS_NM'])['INVC_CONT'].sum()) \
/ md19.groupby(['DL_GD_LCLS_NM'])['INVC_CONT'].sum()

ax, fig = plt.subplots(figsize = (17,7))
cate_inc.plot.bar()

plt.title('2019년 대비 2020년 카테고리별 송장 건수 증가율')
plt.xlabel('상품카테고리', fontsize = 20)
plt.show()
```

# 지역별 송장 건수 변화

```python
fig, ax = plt.subplots(4, 2, figsize = (30,40))

gu = ['서울 중구', '서울 노원구', '대구 중구', '대구 수성구']
for i,j in enumerate(gu):
    md19.query('CTGG_NM == @j').groupby('HDNG_NM').sum().plot.bar(ax = ax[i,0])
    md20.query('CTGG_NM == @j').groupby('HDNG_NM').sum().plot.bar(ax = ax[i,1])
    ax[i,0].set_title(f'2019년 {j}의 동별 송장 건수', fontsize = 20)
    ax[i,1].set_title(f'2020년 {j}의 동별 송장 건수', fontsize = 20)

plt.show()
```

```python
# 각 동별 물류 송장 건수으 증가율을 계산
gu_incr = ((md20.groupby('HDNG_NM').sum() - md19.groupby('HDNG_NM').sum())\
/ md19.groupby('HDNG_NM').sum()).reset_index()
gu_incr = gu_incr.rename(columns={gu_incr['HDNG_NM'].name:'HDONG_NM'})
```

```python
#증가율을 나타낸 데이터프레임과 지리 정보를 병합.
gu_incr = geo.merge(gu_incr, on = 'HDONG_NM')
```

```python
# geodataframe 형식을 시각화
def visualize_gu(data):
    
    fig, ax = plt.subplots(2, 2, figsize=(30, 30))
    
    data.query('SIDO_NM == "서울특별시" and SGNG_NM == "중구"').plot(column='INVC_CONT', cmap = 'Blues', legend=True, ax=ax[0,0])
    ax[0,0].set_title('서울 중구 물류 송장 건수 증가율', fontsize = 25)
    ax[0,0].set_axis_off()
    
    data.query('SIDO_NM == "대구광역시" and SGNG_NM == "중구"').plot(column='INVC_CONT', cmap = 'Blues', legend=True, ax=ax[1,0])
    ax[1,0].set_title('대구 중구 물류 송장 건수 증가율', fontsize = 25)
    ax[1,0].set_axis_off()
  
    data.query('SGNG_NM == "노원구"').plot(column='INVC_CONT', cmap = 'Blues', legend=True, ax=ax[0,1])
    ax[0,1].set_title('서울 노원구 물류 송장 건수 증가율', fontsize = 25)
    ax[0,1].set_axis_off()
    
    data.query('SGNG_NM == "수성구"').plot(column='INVC_CONT', cmap = 'Blues', legend=True, ax=ax[1,1])
    ax[1,1].set_title('대구 수성구 물류 송장 건수 증가율', fontsize = 25)
    ax[1,1].set_axis_off()

    plt.show()  
```

```python
visualize_gu(gu_incr)
```
